```python
df1 = pd.read_excel('construct_validity.xlsx', index_col=0)
df1
```

```python
df1 = df1.sort_values(by='Экономическая дискриминация', ascending = True)
```

```python
df1.plot(x='Рейтинг', 
        kind='bar', 
        stacked=False, 
        figsize=(10,6),
        title='Конструктивная валидность международных гендерных индексов');
```

```python
g = sns.catplot(
    data=df1, kind="bar",
   x="Насилие",  y = "Рейтинг", hue = 'Организация',
     height=8
)
```
